**Задача №1** 

Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество уникальных гео-меток всех пользователей.

*Пример работы программы:*

>ids = {'user1': [213, 213, 213, 15, 213], 

>'user2': [54, 54, 119, 119, 119], 

>'user3': [213, 98, 98, 35]}

>Результат: {98, 35, 15, 213, 54, 119}

```python
ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}

ids_list = [] #создаем список, куда запишем значения из словаря
for items in ids.values(): #через values() возвращаем все значения словаря - списки идентификаторов.
#if type(items) == list: #ОПЦИОНАЛЬНО проверяем, что значения являются списками 
    ids_list += items #добавляем в наш список все значения - списки из словаря, чтобы получить все гео-метки
print(set(ids_list)) #открываем все списки из списка ids_list и удаляем из него дубликаты 
```

**Задача №2**

Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

*Пример работы программы:*
    
>queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

>Результат:

>Поисковых запросов, содержащих 2 слов(а): 42.86%

>Поисковых запросов, содержащих 3 слов(а): 57.14%

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
total_queries = len(queries) #всего запросов
word_counter = {} #пропишем в словаре, сколько запросов состоят из 2 и 3 слов в словаре

for query in queries:
    count = len(query.split())    #считаем количество слов в запросе
    if count not in word_counter: #если статистики по такому кол-ву слов нет,
                                  #фиксируем такой тип запросов в словаре
        word_counter[count] = 0
    word_counter[count] += 1  # увеличиваем счетчик для конкретного типа запроса

for count, number in word_counter.items(): #обращаемся к значению для каждого типа запросов из словаря
                                           #и выводим для них искомую статистику
    answer = (number / total_queries) * 100 
    print(f"Поисковых запросов, содержащих {count} слов(а): {answer:.2f}%")
```
